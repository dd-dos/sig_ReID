```python
import os
img_list = []
for index, folder in enumerate(os.listdir(data_path)):
    temp = []
    folder_path = os.path.join(data_path, folder)
    for img in os.listdir(folder_path):
        if img.lower().endswith('png'):
            img_path = os.path.join(folder_path, img)
            temp.append((img_path, index))
    if len(temp) != 0:
        img_list.append(temp)
```

```python
model = Sig_Ver_Model()
import numpy as np
import random

def check_legit(model, img1_data, img2_data, confusion_dict, threshold=0.3):
    img1_path, label1 = img1_data
    img2_path, label2 = img2_data
    
    pred, _ = model.verify(img1_path, img2_path, threshold)
    truth = (label1==label2)
    
    if pred == truth and truth == True:
        confusion_dict["TP"] += 1
    elif pred == truth and truth == False:
        confusion_dict["TN"] += 1
    elif pred == True and truth == False:
        confusion_dict["FP"] += 1
    else:
        confusion_dict["FN"] += 1
    #print("current acc: {}".format(accuracy(confusion_dict)))
    return confusion_dict

def accuracy(confusion_dict):
    TP, TN, FP, FN = confusion_dict.values()
    return (TP + TN) / (TP + TN + FP + FN)

class BREAK(Exception): pass

def evaluate(model, img_list, threshold=0.3):
    confusion_dict={"TP":0, "TN":0, "FP":0, "FN":0}
    img_list = sum(img_list, [])
    random.shuffle(img_list)
    holder = accuracy(check_legit(model, img_list[0], img_list[1], confusion_dict, threshold))
    for i in range(len(img_list)-1):
        confusion_dict = check_legit(model, img_list[i], img_list[i+1], confusion_dict, threshold)
        current_acc = accuracy(confusion_dict)
        print("#{} current acc {}".format(i, current_acc))
    print("+++++++++++++++++++++++")
    return confusion_dict
```

```python
model = Sig_Ver_Model()
threshold = [i*0.1 for i in range(10)]
acc = [evaluate(model, img_list, threshold[i]) for i in range(10)]
```
